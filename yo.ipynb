```python
def EncodeSex(sex):
    if sex=="male":
        return 1
    else:
        return 0
```

```python
df['Sex']=df['Sex'].apply(EncodeSex)
df
```

```python
def encodeEmbark(embarked):
    if embarked=="S":
        return -1
    elif embarked=="C":
        return 0
    elif embarked=="Q":
        return 1
```

```python
df['Embarked']=df['Embarked'].apply(encodeEmbark)
df
```

```python
from sklearn import model_selection

if __name__ == "__main__":
    df.loc[:, "kfold"] = -1
    df = df.sample(frac = 1).reset_index(drop=True)
    y = df.Survived.values
    skf = model_selection.StratifiedKFold(n_splits=5)
    from sklearn.preprocessing import LabelEncoder
    label_encoder = LabelEncoder()
    y = label_encoder.fit_transform(y)
    from sklearn.utils.multiclass import type_of_target
    print(type_of_target(y))
    
    for f, (t_, v_) in enumerate(skf.split(X=df,y=y)):
        df.loc[v_, "kfold"] = f
        df.to_csv("train_fold.csv")
```

```python
from sklearn import linear_model
from sklearn import metrics
from sklearn.metrics import mean_squared_error
from sklearn.ensemble import AdaBoostClassifier

def run_training(fold):
    df1 = df7
    #df1_train = df1[df.kfold != fold].reset_index(drop = True)
    #df1.kfold
    
    df1_test = df1[df1.kfold == fold].reset_index(drop = True)
    df1_test.drop(['kfold'],axis=1,inplace=True)
   #print(df1_test)
    
    df1_train = df1[df1.kfold != fold].reset_index(drop = True)
 
    df1_train.drop(['kfold'],axis=1,inplace=True)
   

    x_train = df1_train.drop(['Survived'],axis = 1)
    x_valid = df1_train.Survived.values
    y_train = df1_test.drop(['Survived'],axis = 1)
    y_valid = df1_test.Survived.values
    
    
    clf = AdaBoostClassifier(n_estimators=100, random_state=0)
    clf.fit(x_train, x_valid)
    AdaBoostClassifier(n_estimators=100, random_state=0)
    pdt=clf.predict(y_train)
    print(accuracy_score(y_valid,pdt))
    
    
    
    return clf


for i in range(5):
    c=run_training(i)
```
